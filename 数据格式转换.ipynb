```python
import pandas as pd

# 读取 Excel 文件
excel_file = 'data/AI模型性能.xlsx'  # 将此路径替换为实际的文件路径
df = pd.read_excel(excel_file)

# 转换为 JSON 格式
models_json = df.to_dict(orient='records')


# 将 'MMLU' 或 'GSM8K' 为 '/' 的数据删除
models_json = [model for model in models_json if model["MMLU"] != '/' and model["GSM8K"] != '/']


# 替换 'size' 为 NaN 或 0 的值为 1000
for model in models_json:
    if pd.isna(model["参数大小"]) or model["参数大小"] == 0.0:
        model["参数大小"] = 1000

# 打印处理后的结果
for model in models_json:
    print(model)
```
